```python
import matplotlib.pyplot as plt

# Data
delta_t = [10, 20, 30, 40, 50]
upper_bound = [63.93, 63.93, 63.93, 63.93, 63.93]
lower_bound = [63.44, 62.60, 61.81, 60.99, 60.26]
our_method = [63.91, 63.84, 63.78, 63.65, 63.52]

# Create a figure with a transparent background
plt.figure(figsize=(8, 5), facecolor='none')

# Plot each method with distinct styles
plt.plot(delta_t, upper_bound, label='Upper Bound', marker='s', color='blue', linewidth=2.5, markersize=8)
plt.plot(delta_t, lower_bound, label='Lower Bound', marker='o', color='red', linewidth=2.5, markersize=8)
plt.plot(delta_t, our_method, label='Our Method', marker='^', color='green', linewidth=2.5, markersize=8)

# Customize grid and add minor ticks
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.6)
plt.minorticks_on()
plt.grid(which='minor', color='gray', linestyle=':', linewidth=0.5, alpha=0.3)

# Labels and Title
plt.xlabel("Interframe Interval $\Delta t$ (ms)", fontsize=14, labelpad=10, color='black')
plt.ylabel("mIoU (\%)", fontsize=14, labelpad=10, color='black')
# plt.title("Comparison of Upper Bound, Lower Bound, and Our Method", fontsize=16, pad=15, color='black')

# Adjust ticks and legend
plt.xticks(delta_t, fontsize=12, color='black')
plt.yticks(fontsize=12, color='black')
plt.legend(fontsize=12, loc='lower left', frameon=True)

# Adjust axis limits
plt.ylim(59, 65)
plt.xlim(0, 55)

# Save as a transparent PNG file
plt.tight_layout()
plt.savefig("shf_dsec_results_transparent_fixed.png", dpi=300, transparent=True)
plt.show()
```
