```python
import matplotlib.pyplot as plt
import re

# 获取所有类名
all_class_names = [
    'background', 'building', 'fence', 'person', 'pole', 
    'road', 'sidewalk', 'vegetation', 'car', 'wall', 'traffic sign'
]

def parse_log_file(file_path):
    results = {}
    results['mIoU'] = {}
    for class_name in all_class_names:
        results[class_name] = {}

    with open(file_path, 'r') as file:
        lines = file.readlines()

    for i, line in enumerate(lines):
        # 提取 epoch 和 mean_IoU
        # epoch_match = re.search(r'Current epoch:(\d+)\s+mIoU:\s+([\d.]+)', line)
        # Current epoch:20 mIoU: 56.44 Best mIoU: 56.44
        epoch_match = re.search(r'Current epoch:(\d+)\s+mIoU:\s+([\d.]+)\s+Best mIoU:\s+([\d.]+)', line)
        if epoch_match:
            epoch = int(epoch_match.group(1))
            mean_iou = float(epoch_match.group(2))
            best_miou = float(epoch_match.group(3))
            results['mIoU'][epoch] = mean_iou
            results['Best_mIoU'] = best_miou


        # 提取每一类的结果
        class_match = re.match(r'(\d+)\s+([\w\s]+)\s+([\d.]+)\s+([\d.]+)', line)
        if class_match:
            class_id = int(class_match.group(1))
            class_name = class_match.group(2).strip()  # 去除多余的空格
            iou = float(class_match.group(3))
            acc = float(class_match.group(4))
            epoch_line_index = i + (12 - class_id)

            if epoch_line_index < len(lines):
                epoch_line = lines[epoch_line_index]
                epoch_match = re.search(r'== Epoch:\s+(\d+)\s+mean_IoU\s+([\d.]+)', epoch_line)
                if epoch_match:
                    epoch = int(epoch_match.group(1))
                    results[class_name][epoch] = {'iou': iou, 'acc': acc}

    return results

# Define the file path
results_dicts = {}
import os
file_dir = '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/'
# file_dir下所有以rgb_day_11_2024开头的文件
# file_paths = sorted(os.path.join(file_dir, file) for file in os.listdir(file_dir) if file.startswith('rgb_day_11_2024'))
file_paths = [
    # train from scratch t0-t0 67.67
    '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_night_11_20241110_213215_train.log',
    # train from scratch t0-t1 67.67
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_035802_train.log',
    # train from scratch t0-t2 62.42
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_035441_train.log',
    # eraft flow 50ms event_20
    '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_night_11_20241110_213411_train.log',
    # raft flow 50ms event_20
    '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_night_11_20241110_215413_train.log',
    '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_night_11_20241111_032118_train.log',  # pretrained on day 73.99
    # zero flow 100ms event_20
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_160704_train.log',  # 67.88
    # eraft flow 100ms event_40 塌了
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_194519_train.log',
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241031_023135_train.log',  # iterative bs=9
    # eraft flow 50ms event_20 72.67
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_113014_train.log',  # 72.67
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_181251_train.log',  # clip norm 73.12 # NOTE
    # raft flow 100ms event_20 71.30
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_180305_train.log',
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_183957_train.log',    # 71.30
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_212702_train.log',  # clip norm 71.39
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_224508_train.log',  # clip norm & iterative 2 没啥用
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241031_010021_train.log',  # clip norm & iterative 2 debug 没啥用
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_001957_train.log',  # new machine test
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_003742_train.log',  # new machine test
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_010340_train.log',  # new machine test
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_012122_train.log',  # new machine test

    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_035304_train.log',  # new machine test# -50-0 0-50 eraft 73.3
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_035414_train.log',  # new machine test# -100-0 0-100 eraft 71.95
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_143543_train.log',  # new machine test# -50-0 0-50 50-100 eraft
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_143607_train.log',  # new machine test# -50-0 0-50 50-100 eraft with mid supervised
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_174219_train.log',  # new machine test# -50-0 0-50 50-100 eraft with flow1+flow2
    # '/home/xy/work/adap_v/DELIVER/output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_165951_train.log',  # new machine test# -50-0 0-50 50-100 eraft with [event2, img2] in cnet
]
for file_path in file_paths:
    parts = file_path.split('/')
    filename = parts[-1]
    # scene = filename.split('_')[:2]
    # scene = '_'.join(scene)
    # print(scene)

    # Initialize an empty dictionary to store the results
    results_dict = parse_log_file(file_path)
    results_dicts[filename] = results_dict
```

```python
def compute_iou(hist):
    ious = hist.diag() / (hist.sum(0) + hist.sum(1) - hist.diag())
    ious[ious.isnan()]=0.
    miou = ious.mean().item()
    # miou = ious[~ious.isnan()].mean().item()
    ious *= 100
    miou *= 100
    return ious, miou
```
